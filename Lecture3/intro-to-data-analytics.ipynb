# Numpy

*  Basic difference between Python list and NumPy is that Numy arrays are more efficient for numerical operations
  

```python
python_list = [1,2,3,4,5,6]

np_array = np.array(python_list)

print(np_array)

print("Python list * 3:", python_list * 3)
print("Numpy array * 3:", np_array * 3)

print(np_array ** 2)

print("Array shape: ", np_array.shape)
print("Mean of array:", np_array.mean())

arr = np_array.reshape(2, 3)
print("Reshaped array: ", arr)

print("Transpose array: ", arr.T)

print("Access sub array: ", arr[1:2, 0:2])
```

```python
# Filtering using masks

mask = arr > 2

print("Filtered array: ", arr[mask])

mask2 = arr % 2 == 0

print("filtered array: ", arr[mask2])


# Adding arrays using Broadcasting

arr1 = np.array([1,2,3])

# Create a 2D array
arr2 = np.array([[4], [5], [3]])

# Add arrays using broadcasting

result = arr1 + arr2 

print("Array 1: \n", arr1)
print("Array 2: \n", arr2)
print("Result of broadcasting:\n", result)
```

# Loading the Netflix Titles dataframe

```python
# Load the dataset: Use Pandas to read the CSV file

df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

# peek at data
df.head(10)

#df.tail()

#print(df)
```

```python
# how to access a single column
#print(df[['title', 'country']].head(10))

# Find missing values: Check how many values are missing in each column
df.isnull().sum()

# handle missing data
# Fill missing categorical data with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df.isnull().sum()
```

# Filtering and Transforming Data

```python
# Filter for TV shows produced in India

indian_shows = df[(df['country'].str.contains('India') & (df['type'] == 'TV Show')) ]

indian_shows.shape
indian_shows.head()
```

```python
#Filter for movies released after 2015

recent_movies = df[(df['type']  == 'Movie') & (df['release_year'] > 2015)]

recent_movies.shape
recent_movies.head()
```

```python
# Count the number of movies and TV shows using value_counts()
movies_tv_count = df['type'].value_counts()
print("Movies vs TV shows: ", movies_tv_count)

country_count = df['country'].value_counts()
print(country_count)
```

```python
# NumPy operations on DF

years = np.array(df['release_year'])

print("Mean release year: ", np.mean(years))
print("Standard Deviation: ", np.std(years))
```
